```python
import math
```

```python
class calc():
    def __init__(self, input1, input2, input3):
        self.input1=input1
        self.input2=input2
        self.input3=input3
    
    def add(num1, num2):
        return f"addition= {num1+num2}"
    
    def subt(num1, num2):
        return f"subtraction= {num1-num2}"
    
    def mult(num1, num2):
        return f"multiplication= {num1*num2}"
    
    def div(num1, num2):
        return f"division= {num1/num2}"
    
    def exp(num1, num2):
        return f"{num1} raised to the power of {num2} = {num1**num2}"
    
    def sq_rt(num1):
        if num1>=0:
            return f"square root= {num1**(1/2)}"
        else:
            return f"The number you have chosen is a negative number"
        
    def cube_rt(num1):
        return f"cube root= {num1**(1/3)}"
    
    def other_rt(num1, num2):
        return f"root= {num2**(1/num1)}"
    
    def nat_log(num1):
        return f"natural log= {math.log(num1)}"
    
    def sin(num1):
        return f"sine of {num1}= {math.sin(num1)}"
    
    def cos(num1):
        return f"cosine of {num1}= {math.cos(num1)}"
        
    def tan(num1):
        return f"tangent of {num1}= {math.tan(num1)}"
    
    def min1(num1, num2, num3):
        return f"minimum= {min(num1, num2, num3)}"
        
    def max1(num1, num2, num3):
        return f"maximum= {max(num1, num2, num3)}"
    
    def log_e(num1):
        return f"exponential value with base e = {math.exp(num1)}"

while True:
    print("\nChoose from the following menu:")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Exponent")
    print("6. Square Root")
    print("7. Cube Root")
    print("8. Other Roots")
    print("9. Natural Log")
    print("10. Sine Function")
    print("11. Cosine Function")
    print("12. Tangent Function")
    print("13. Minimum")
    print("14. Maximum")
    print("15. Exponential Value with base e")
    print("16. Exit")
    
    choice = input("Enter your choice (1-16): ")
    
    if choice=="1":
        num1 = int(input("Enter the first number: "))
        num2 = int(input("Enter the second number: "))
        print(calc.add(num1, num2))
    elif choice=="2":
        num1 = int(input("Enter the first number: "))
        num2 = int(input("Enter the second number: "))
        print(calc.subt(num1, num2))
    elif choice=="3":
        num1 = int(input("Enter the first number: "))
        num2 = int(input("Enter the second number: "))
        print(calc.mult(num1, num2))
    elif choice=="4":
        num1 = int(input("Enter the first number: "))
        num2 = int(input("Enter the second number: "))
        if num2==0:
            print("Infinity")
        else:
            print(calc.div(num1, num2))
    elif choice=="5":
        num1 = int(input("Enter the base number: "))
        num2 = int(input("Enter the exponent: "))
        print(calc.exp(num1, num2))
    elif choice=="6":
        num1 = int(input("Enter the number: "))
        print(calc.sq_rt(num1))
    elif choice=="7":
        num1 = int(input("Enter the number: "))
        print(calc.cube_rt(num1))
    elif choice=="8":
        num1= int(input("Enter the custom root: "))
        num2= int(input("Enter the base number: "))
        print(calc.other_rt(num1,num2))
    elif choice=="9":
        num1= int(input("Enter the number: "))
        print(calc.nat_log(num1))
    elif choice=="10":
        num1= int(input("Enter the number: "))
        print(calc.sin(num1))
    elif choice=="11":
        num1= int(input("Enter the number: "))
        print(calc.cos(num1))
    elif choice=="12":
        num1= int(input("Enter the number: "))
        print(calc.tan(num1))
    elif choice=="13":
        num1= int(input("Enter the first number: "))
        num2= int(input("Enter the second number: "))
        num3= int(input("Enter the third number: "))
        print(calc.min1(num1,num2,num3))
    elif choice=="14":
        num1= int(input("Enter the first number: "))
        num2= int(input("Enter the second number: "))
        num3= int(input("Enter the third number: "))
        print(calc.max1(num1,num2,num3))
    elif choice=="15":
        num1= int(input("Enter the number: "))
        print(calc.log_e(num1))
    elif choice=="16":
        break
    else:
        print("Not a valid choice")
```
